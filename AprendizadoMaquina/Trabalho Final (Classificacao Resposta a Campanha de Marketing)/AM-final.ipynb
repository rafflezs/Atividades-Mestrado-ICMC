# Análise Exploratória
## Pré-processamento

```python
import pandas as pd

init_df = pd.read_csv("data/marketing_campaign.csv", delimiter="\t")

display(init_df.sample(5))
```

```python
print(f"O dataframe tem dimensao {init_df.shape}")
print(f"Existem {init_df.duplicated().sum()} valores duplicados no dataset")
display(init_df.describe())
```

```python
display(init_df.isna().sum())
```

```python
init_df.dtypes
```

```python
for cat_feat in ["Education", "Marital_Status"]:
    init_df[cat_feat] = pd.Categorical(init_df[cat_feat])
    init_df[cat_feat+"_Code"] = pd.Categorical(init_df[cat_feat]).codes

display(init_df[["Education", "Marital_Status", "Education_Code", "Marital_Status_Code"]])
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
sns.histplot(init_df["Z_CostContact"], ax=axes[0])
sns.histplot(init_df["Z_Revenue"], ax=axes[1])
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
sns.histplot(init_df["AcceptedCmp1"], ax=axes[0])
sns.histplot(init_df["AcceptedCmp3"], ax=axes[1])
fig.tight_layout()
```

```python
init_df.drop(columns=["Z_CostContact", "Z_Revenue"], inplace=True)
init_df["Dt_Customer"] = pd.to_datetime(init_df["Dt_Customer"], dayfirst=True)
init_df["Age"] = init_df["Year_Birth"].apply(lambda x: 2024 - x)
```

### Mapeando dados categóricos - Cheatsheet

```python
education_cheatsheet = pd.DataFrame({
    "Education": init_df["Education"].cat.categories,
    "Education_Code": range(len(init_df["Education"].cat.categories))
})

marital_status_cheatsheet = pd.DataFrame({
    "Marital_Status": init_df["Marital_Status"].cat.categories,
    "Marital_Status_Code": range(len(init_df["Marital_Status"].cat.categories))
})

display(education_cheatsheet)
display(marital_status_cheatsheet)
```

```python
init_df = init_df.drop(columns=["Education", "Marital_Status"])
df = init_df.copy()
display(df)
del init_df
```
